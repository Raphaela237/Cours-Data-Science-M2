# Pandas for Data Science
This notebook demonstrates powerful capabilities of the Pandas library using a complex dataset.


```python
import pandas as pd
import numpy as np
# load employee data
df = pd.read_csv('employee_data.csv')
# Display the first few rows of the dataset
df.head()
```

```python
# Average salary and performance by Department and Team
agg_df = df.groupby(['Department', 'Team']).agg({
    'Salary': ['mean', 'max', 'min'],
    'PerformanceScore': ['mean', 'std'],
    'Age': 'median'
})
agg_df
```

```python
# Employees in Engineering, older than 40, with high performance
filtered = df[
    (df['Department'] == 'Engineering') & 
    (df['Age'] > 40) & 
    (df['PerformanceScore'] > 3.5)
]
filtered.sort_values(by='PerformanceScore', ascending=False)
```

```python
df['JoinDate'] = pd.to_datetime(df['JoinDate'])
```

```python
def experience_level(row):
    years = 2025 - row['JoinDate'].year
    if years < 2:
        return 'Junior'
    elif years < 5:
        return 'Mid'
    else:
        return 'Senior'

df['ExperienceLevel'] = df.apply(experience_level, axis=1)
df[['Name', 'JoinDate', 'ExperienceLevel']].head()
```

```python
pivot = pd.pivot_table(df, 
                         index='Department', 
                         columns='ExperienceLevel', 
                         values='Salary', 
                         aggfunc='mean')
pivot
```

```python
# Simulate a new DataFrame with bonus info
bonus_df = pd.DataFrame({
    'EmployeeID': np.random.choice(df['EmployeeID'], size=50, replace=False),
    'Bonus': np.random.randint(1000, 10000, size=50)
})

# Merge with main data
df_merged = df.merge(bonus_df, on='EmployeeID', how='left')
df_merged[['Name', 'Department', 'Bonus']].head()
```

```python
# Monthly new joiners
monthly_joins = df.set_index('JoinDate').resample('M')['EmployeeID'].count()
monthly_joins.plot(title='Monthly New Joiners', figsize=(10,4))
```

## Exercice: Sales Data Analysis

You are provided with a dataset `sales_data.csv` containing retail transactions with the following columns:

- `OrderID`: Order identifier  
- `CustomerID`: Unique customer identifier  
- `Product`: Product name  
- `Category`: Product category  
- `Quantity`: Number of units sold  
- `Price`: Unit price  
- `OrderDate`: Date of the order  
- `Country`: Customer's country  

### Tasks:

1. **Basic Exploration**:
   - Display the first 5 rows.
   - Count missing values in each column.
   - Print summary statistics for `Quantity` and `Price`.

2. **Data Cleaning**:
   - Remove rows where `Quantity` or `Price` is less than or equal to 0.
   - Convert `OrderDate` to datetime format.

3. **Feature Engineering**:
   - Create a new column `TotalAmount` = `Quantity` × `Price`.
   - Extract the `Month` and `DayOfWeek` from `OrderDate`.

4. **Analysis**:
   - What is the total revenue per `Country`? (Sort descending)
   - What are the top 5 best-selling products by `Quantity`?
   - How many unique customers are there in each `Country`?

5. **Time Series Aggregation**:
   - Plot total daily revenue (`OrderDate` vs. `TotalAmount`) using a line plot.
   - Compute the monthly average quantity sold per category.

> 💡 Use `.groupby()`, `.agg()`, `.pivot_table()`, and visualization functions where appropriate.

```python
# with sales dataset
# Chargement du fichier
df_sales = pd.read_csv("sales_data.csv")

# Afficher les 5 premières lignes
df_sales.head()

```

```python
# Compter les valeurs manquantes
print(df_sales.isnull().sum())

# Statistiques de base pour Quantity et Price
df_sales[['Quantity', 'Price']].describe()
```

## Nettoyage des données

```python
# Supprimer les lignes où Quantity ou Price sont <= 0
df_sales = df_sales[(df_sales['Quantity'] > 0) & (df_sales['Price'] > 0)].copy()

# Convertir OrderDate en datetime
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])
```

## Feature Engineering

```python
# Créer TotalAmount
df_sales['TotalAmount'] = df_sales['Quantity'] * df_sales['Price']

# Extraire le mois et le jour de la semaine
df_sales['Month'] = df_sales['OrderDate'].dt.month
df_sales['DayOfWeek'] = df_sales['OrderDate'].dt.day_name()
```

```python
df_sales.head()
```

### Revenu total par pays (descendant)

```python
revenue_by_country = df_sales.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False)
print(revenue_by_country)
```

### Top 5 des produits les plus vendus (en quantité)

```python
top_products = df_sales.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

### Nombre de clients uniques par pays

```python
unique_customers = df_sales.groupby('Country')['CustomerID'].nunique()
print(unique_customers)
```

## Agrégations temporelles
### Courbe du revenu quotidien


```python
import matplotlib.pyplot as plt

daily_revenue = df_sales.groupby('OrderDate')['TotalAmount'].sum()

plt.figure(figsize=(12, 5))
daily_revenue.plot(kind='line')
plt.title("Revenu total par jour")
plt.xlabel("Date")
plt.ylabel("TotalAmount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Moyenne mensuelle des quantités par catégorie

```python
monthly_avg_quantity = df_sales.groupby(['Category', 'Month'])['Quantity'].mean().reset_index()

# tableau croisé pour meilleure lisibilité
pivot_table = pd.pivot_table(df_sales, values='Quantity', index='Month', columns='Category', aggfunc='mean')
print("\nMonthly average quantity sold per category:")
print(pivot_table)
```

```python
import seaborn as sns
# Plot monthly average quantity sold per category
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_avg_quantity, x='Month', y='Quantity', hue='Category', marker='o')
plt.title('Monthly Average Quantity Sold per Category')
plt.xlabel('Month')
plt.ylabel('Average Quantity Sold')
plt.xticks(range(1,13))
plt.grid(True)
plt.tight_layout()
plt.show()
```
